```python
import pandas as pd

# 1. Скачивание
url = "https://www.policyuncertainty.com/media/Russia_Policy_Uncertainty_Data.xlsx"
df_raw = pd.read_excel(url)

# 2. Дропаем последнюю строку (если там пустые значения или технические подписи)
df_raw = df_raw.iloc[:-1]

# 3. Приведение даты к формату YYYY-MM
# В этом файле колонки называются 'Year' и 'Month'
df_raw['Date'] = pd.to_datetime(df_raw[['Year', 'Month']].assign(Day=1))
df_raw['Date_YYYY_MM'] = df_raw['Date'].dt.to_period('M')

# 4. Формируем финальный вид
target_col = 'News-Based Policy Uncertainty Index'
mpu_index = df_raw[['Date_YYYY_MM', target_col]].copy()
mpu_index.columns = ['Date', 'MPU_Index']

print("Размер таблицы после удаления последней строки:", mpu_index.shape)
print(mpu_index.tail())
```

```python
mpu_index.to_csv('RussiaNewsBasedUncertainty.csv', index=False)
mpu_index.to_excel('RussiaNewsBasedUncertainty.xlsx', index=False)
```
