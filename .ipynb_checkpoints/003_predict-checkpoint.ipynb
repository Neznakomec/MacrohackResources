```python
import numpy as np
import pandas as pd
```

```python
mpu = pd.read_csv('mpu_index.csv')
```

```python
mpu
```

```python
data = pd.read_csv('cbr_key_rate.csv')
data['Дата'] = pd.to_datetime(data['Дата'])
```

```python
df = data
g = df.groupby(pd.Grouper(freq='ME', key='Дата'))
df_monthly = df[df['Дата'].eq(g['Дата'].transform('max'))]
df_monthly
```

```python
def rolling_realized_volatility(series, window=22):
    """
    Расчет волатильности в скользящем окне.
    window: размер окна (например, 22 рабочих дня)
    """
    log_returns = np.log(series / series.shift(1))
    
    # Считаем корень из скользящей суммы квадратов
    rolling_rv = np.sqrt((log_returns**2).rolling(window=window).sum())
    return rolling_rv

# Пример для цен:
df = pd.DataFrame(df_monthly)
df['RV_Rolling_10'] = rolling_realized_volatility(df['Ключевая_ставка'], window=10)

print(df.tail())
```

```python
df['RV_Growth_1m'] = df.diff(1)['RV_Rolling_10']
df['RV_Growth_3m'] = df.diff(3)['RV_Rolling_10']
df
```

```python
df
```

```python
# 1. Убеждаемся, что обе колонки имеют тип datetime
df['Дата'] = pd.to_datetime(df['Дата'])
mpu['Date'] = pd.to_datetime(mpu['Date'])

# 2. Создаем временные колонки для объединения (формат 'YYYY-MM')
# Используем dt.to_period('M'), это самый надежный способ для стыковки месяцев
df['month_key'] = df['Дата'].dt.to_period('M')
mpu['month_key'] = mpu['Date'].dt.to_period('M')

# 3. Объединяем (merge)
# how='inner' оставит только те месяцы, которые есть в обоих файлах
# how='left' оставит все строки из основной таблицы со ставкой
merged_df = pd.merge(df, mpu[['month_key', 'MPU_Index']], on='month_key', how='inner')

# 4. Удаляем вспомогательную колонку, если она больше не нужна
merged_df = merged_df.drop(columns=['month_key'])

# Проверяем результат
print(merged_df.head())
```

```python
merged_df
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. Очистка данных от NaN (они возникают из-за rolling и shift)
final_df = merged_df[['RV_Growth_1m', 'MPU_Index']].dropna()

# 2. Определение целевой переменной (y) и предиктора (X)
# X должен быть двумерным массивом
X = final_df[['MPU_Index']] 
y = final_df['RV_Growth_1m']

# 3. Создание и обучение модели
model = LinearRegression()
model.fit(X, y)

# 4. Вывод результатов
intercept = model.intercept_
slope = model.coef_[0]
r_sq = model.score(X, y)

print(f"Результаты регрессии:")
print(f"Уравнение: RV_Growth_1m = {slope:.6f} * MPU_Index + {intercept:.6f}")
print(f"Коэффициент детерминации (R²): {r_sq:.4f}")

# 5. Визуализация
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.5, label='Фактические данные')
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Линия регрессии')
plt.xlabel('MPU_Index')
plt.ylabel('RV_Growth_1m')
plt.title('Регрессия реализованной волатильности по индексу MPU')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
