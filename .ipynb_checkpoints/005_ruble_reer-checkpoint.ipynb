```python
import pandas as pd
import numpy as np

# 1. Прямая ссылка на скачивание CSV файла с FRED
# Тикер RBRUBIS — это Real Effective Exchange Rate для России
url = "https://fred.stlouisfed.org/graph/fredgraph.csv?id=RBRUBIS"

# 2. Чтение данных напрямую по ссылке
# FRED отдает CSV с колонками 'DATE' и 'RBRUBIS'
reer = pd.read_csv(url)

# 3. Приводим названия колонок к вашему стандарту
reer.columns = ['Date', 'RBRUBIS']

# 4. Очистка и преобразование данных
reer['Date'] = pd.to_datetime(reer['Date'])

# Превращаем значение индекса в число (errors='coerce' удалит "точки" или пустые строки, если они есть)
reer['RBRUBIS'] = pd.to_numeric(reer['RBRUBIS'], errors='coerce')
reer = reer.dropna()

# 5. Форматирование даты в YYYY-MM
# Важно: после этого колонка 'Date' станет строковой, что удобно для CSV
reer['Date'] = reer['Date'].dt.strftime('%Y-%m')

# 6. Сохранение в CSV
reer.to_csv('reer_russia.csv', index=False)

# 7. Сохранение в Excel
reer.to_excel('reer_russia.xlsx', index=False)

# Вывод результата для проверки
print("Данные успешно скачаны и сохранены!")
print(reer.tail())
```
